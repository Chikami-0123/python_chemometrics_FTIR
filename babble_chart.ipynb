```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'X': [10, 20, 30, 40, 50],
    'Y': [25, 15, 35, 45, 55],
    'Size': [100, 200, 300, 400, 500]
}

# pandas DataFrameにデータを変換します。
df = pd.DataFrame(data)

# グラフのスタイルを設定します。
plt.style.use('ggplot')

# バブルチャートを作成します。
scatter = plt.scatter(df['X'], df['Y'], s=df['Size'], alpha=0.5)

# 軸ラベルを設定します。
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# グラフのタイトルを設定します。
plt.title('Bubble Chart Example')

# 凡例を表示します。
for i, txt in enumerate(df['Category']):
    plt.annotate(txt, (df['X'][i], df['Y'][i]))

# グラフを表示します。
plt.show()
```
